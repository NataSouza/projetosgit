```python
# CONECÇÃO COM MYSQL 

import mysql.connector

def insere_usuario(con, ORDER_ID, ORDER_DATE, ORDER_STATUS, DELIVERY_DATE, SELLER_ID, BUYER_ID):
    cursor = con.cursor()
    sql = "INSERT INTO ORDERS (ORDER_ID, ORDER_DATE, ORDER_STATUS, DELIVERY_DATE, SELLER_ID, BUYER_ID) values (%s, %s, %s, %s, %s, %s)"
    valores = (ORDER_ID, ORDER_DATE, ORDER_STATUS, DELIVERY_DATE, SELLER_ID, BUYER_ID)
    cursor.execute(sql, valores)
    cursor.close()
    con.commit()
    
def main():
    con = mysql.connector.connect(host= "localhost", user="conMySQL", password="admin", database="projeto")
    
    insere_usuario(con, "O00001", "2020-12-08", "OK", "2020-12-20", "U00614", "U00450")
    
    con.close()
    
if __name__ == "__main__":
    main()
```
