```python
#ABRIR ARQUIVO XLSX


import pandas as pd
from ast import literal_eval

#DATAFRAME ORDERS
df_Orders = pd.read_excel(r'Orders.xlsx', engine = 'openpyxl')
df_Orders = df_Orders.replace('\n',' ', regex=True)
```

```python
#DATAFRAME PRODUCTS
df_Products = pd.read_excel(r'Products.xlsx', engine = 'openpyxl')
```

```python
#DATAFRAME PRODUCT QUANTITIES
df_product_quanties = pd.DataFrame()
df_product_quanties['A'] = df_Orders['order_id']
df_product_quanties['B'] = df_Orders['product_quantities']
```

```python
#SEPARAR POR DELIMITADOR
df_temp = df_product_quanties['B'].str.split(',', expand=True)
df_temp = df_temp.replace('"','', regex=True)
df_temp = df_temp.replace('{','', regex=True)
df_temp = df_temp.replace('}','', regex=True)

# INSERIR NOVA COLUNA
df_temp['A'] = df_product_quanties['A']

#ALTERAR POSIÇÃO DA COLUNA
fc = df_temp.pop('A')

df_temp.insert(0, 'A', fc)
```

```python
# DANDO O MELT
df_teste = df_temp.melt(id_vars = 'A')
```

```python
# EXCLUINDO COLUNA
df_teste = df_teste.drop(columns=['variable'])
```

```python
# CONCATENAR COLUNAS
df_novo = df_teste['A'].str.cat(df_teste[['value']], sep = ':')
df_novo = pd.DataFrame(df_novo)
```

```python
# SEPARAR COLUNA POR DELIMITADOR
df_novo = df_novo['A'].str.split(':', expand = True)
```
