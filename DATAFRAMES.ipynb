```python
#CADASTRO LOJA A

cadastro_a = {'Id': ['AA2930', 'BB4563', 'CC2139', 'DE2521', 'GT3462', 'HH1158'],
              'Nome': ['Victor', 'Amanda', 'Bruna', 'Carlos', 'Ricardo', 'Maria'],
              'Idade': [20,35,40,54,30,27],
              'CEP': ['00902-029', '11111-111', '22222-888', '00000-999', '88888-111', '77777-666']
             }
cadastro_a = pd.DataFrame(cadastro_a, columns = ['Id', 'Nome', 'Idade', 'CEP'])
cadastro_a
```

```python
#CADASTRO LOJA B

cadastro_b = {'Id': ['CC9999', 'EF4488', 'DD9999', 'GT3462', 'HH1158'],
              'Nome': ['Marcos', 'Patricia', 'Ericka', 'Ricardo', 'Maria'],
              'Idade': [19,30,22,30,27],
              'CEP': ['00902-029', '11111-111', '11111-888', '88888-111', '77777-666']
             }
cadastro_b = pd.DataFrame(cadastro_b, columns = ['Id', 'Nome', 'Idade', 'CEP'])
cadastro_b
```

```python
#REGISTRO DE COMPRAS DE TODA UNIDADES

compras = {'Id': ['AA2930', 'EF4488', 'CC2139', 'EF4488', 'CC999', 'AA2930', 'HH1158', 'HH1158'],
           'Data': ['2019-01-01', '2019-01-30', '2019-01-30', '2019-02-01', '2019-02-20', '2019-03-15', '2019-04-01', '2019-04-10'],
           'Valor': [200,100,40,150,300,25,50,500]          
          }
compras = pd.DataFrame(compras, columns = ['Id', 'Data', 'Valor'])
compras
```

```python
# pd.merge(tabela_da_esquerda, tabela_da_direita, on='coluna_coincidente', how='left|rigth|inner|outer')
```

FUNÇÃO MERGE

É a função que realiza o papel do 'join' no SQL

OBS: O x é referente a tabela da direita e o Y é referente a tabela da direita


```python
#INNER JOIN

pd.merge(cadastro_a, cadastro_b, on=['Id'], how='inner')
```

```python
pd.merge(cadastro_a[['Id']], cadastro_b,  on=['Id'], how='inner')
```

```python
pd.merge(cadastro_a[['Id']], cadastro_b[['Id', 'Idade', 'CEP']],  on=['Id'], how='inner', suffixes =('_A', '_B'))
```

```python
# FULL JOIN

lojas = pd.concat([cadastro_a, cadastro_b], ignore_index=True)

lojas.drop_duplicates(subset="Id")

lojas
```

```python
# pd.merge(tabela_da_esquerda, tabela_da_direita, on='coluna_coincidente', how='left|rigth|inner|outer')

esquerda = pd.merge(cadastro_a, compras, on='Id', how='left')
```

GROUP BY

COMO UTILIZAR O GROUP BY com o HEAD

```python
esquerda.head(4).groupby(['Id', 'Nome', 'Idade'])['Valor'].sum()
```

OUTER JOIN

Realiza a junção de todos os dados

```python
#OUTER JOIN

pd.merge(cadastro_a, cadastro_b, on='Id', how='outer', indicator = True)
```

## AULA 30 pivot_table

EX:

OBS: Se eu não incluir o aggfunc = 'sum', ele vai trazer a media

```python
Carros = [7, 4, 3, 2, 8]
dias = pd.date_range('20190101','20190101', periods=5)
vendedor = ['George', 'Vagner', 'Pedro', 'Vagner', 'George']

df = pd.DataFrame({'Vendas':Carros, 'Data':dias, 'Vendedor':vendedor})
df
```

## AULA EXTRA SEPARAR POR COLUNAS

```python
data = ['São Paulo/SP', 'Brasilia/DF', 'Goiania/GO', 'Rio de Janeiro/RJ', 'Rio Branco/AC', 'Manaus/AM', 'Maceio/AL',
'Belem/PA', 'Curitiba/PR']

df2 = pd.DataFrame(data, index= list(range(50, 59)), columns= ['Local']) 
df2
```

```python
df2[['Cidade', 'Estado']] = df2['Local'].str.split('/', expand = True)
df2
```

```python
datas = ['São Paulo/SP/Brasil', 'Brasilia/DF/Brasil', 'Goiania/GO/Brasil', 'Rio de Janeiro/RJ/Brasil', 'Rio Branco/AC/Brasil', 'Manaus/AM/Brasil', 'Maceio/AL/Brasil',
'Belem/PA/Brasil', 'Curitiba/PR/Brasil']

df3 = pd.DataFrame(datas, columns=['Local'])
df3
```

```python
df3[['Cidade', 'Estado']] = df3['Local'].str.split('/', expand = True, n=1)
df3
```

```python
df3[['Cidade', 'Estado']] = df3['Local'].str.rsplit('/', expand = True, n=1)
df3
```
